```python
import numpy as np
import math
from math import sqrt
from qiskit import ClassicalRegister, QuantumRegister, QuantumCircuit
from qiskit import execute, Aer, BasicAer
from qiskit.tools.visualization import circuit_drawer
from qiskit.tools.visualization import plot_histogram
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
from qiskit import transpile
```

```python
qc = QuantumCircuit(3)        
qc.h(0)
qc.h(1)
qc.h(2)
qc.barrier()   
#qc.h([0,1,2])
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc
psi_1=Statevector(qc)
psi_1.draw('latex')
```

```python
qc.x(0)
qc.cx(1,2)
qc.y(2)
qc.barrier()
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc
psi_2=Statevector(qc)
psi_2.draw('latex')
```

```python
qc.ccx(0,1,2)
qc.x(0)
qc.y(1)
qc.barrier()
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc
psi_3=Statevector(qc)
psi_3.draw('latex')
```

```python
qc.z(1)
qc.swap(0,2) 
qc.barrier()
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc
psi_4=Statevector(qc)
psi_4.draw('latex')
```

```python
qc.barrier()
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc
psi_5=Statevector(qc)
psi_5.draw('latex')
```

```python
qc. measure_all()             # Pengukuran pada seluruh qubit
qc.draw('mpl')                # Tampilkan gambar sirkuit di layar
```

```python
# Simulasikan di komputer lokal
simulator = BasicAer.get_backend("qasm_simulator")
job = execute(qc, simulator, shots = 1000)
result = job.result()
count = result.get_counts()
plot_histogram(count)
```
